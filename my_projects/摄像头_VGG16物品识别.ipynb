```python
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.applications.vgg16 import VGG16
from keras import backend as K
```

```python
'''摄像头按'z'或'Z'catch,任意键结束'''

def get_photo(go_to_file):
    '''传入图片文件名，可为空文件，抓拍后重写文件'''

    cv2.namedWindow("frame")      # 创建窗口显示图片，可加入cv2.……参数调整，见saveFace1.py
    cap = cv2.VideoCapture(0)     # 打开摄像头，内置为0
    while(cap.isOpened()):        # 是否打开
        ret,img = cap.read()      # 读取摄像头图片，读取成功为True
        if ret == True:
            cv2.imshow("frame",img)   # 图片显示在窗口中
            k = cv2.waitKey(1)       # 每1毫秒捕获一次键盘输入，影响画面
            if k == ord("z") or k == ord("Z"):
                cv2.imwrite(go_to_file,img)    # 保存到路径中
                break

    cap.release()    # 关闭摄像头
    cv2.waitKey(0)   # 等待n毫秒后,按任意键进行下面操作，0为无限长
    cv2.destroyWindow("frame")   # 关闭窗口
    return go_to_file    # 返回路径名

get_photo('1.jpg')

K.clear_session()

# Note that we are including the densely-connected classifier on top;
# all previous times, we were discarding it.
model = VGG16(weights='imagenet')  # 注意，网络中包括了密集连接分类器。在前面所有的例子中，我们都舍弃了这个分类器

# The local path to our target image


# `img` is a PIL image of size 224x224，原始的图片大小
img = image.load_img('1.jpg', target_size=(224, 224))
plt.imshow(img)

# `x` is a float32 Numpy array of shape (224, 224, 3)
x = image.img_to_array(img)

# We add a dimension to transform our array into a "batch"
# of size (1, 224, 224, 3)
x = np.expand_dims(x, axis=0)

# Finally we preprocess the batch
# (this does channel-wise color normalization),对批量进行预处理（按通道进行颜色标准化）RGB->BGR再减平均数
x = preprocess_input(x)

preds = model.predict(x)
print('Predicted:', decode_predictions(preds, top=3)[0][0][1])
```
